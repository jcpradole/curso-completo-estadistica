# Graficando datos

```python
df_comp.spx.plot(figsize=(20,5), title = "S&P500 Prices")
plt.show()
```

```python
df_comp.spx.plot(figsize=(20,5), title = "S&P500 Prices")
df_comp.ftse.plot(figsize=(20,5), title = "FTSE100 Prices")
df_comp.dax.plot(figsize=(20,5), title = "DAX100 Prices")
plt.title("S&P vs FTSE vs DAX")
plt.show()
```

# The QQ Plot: Quantile Quantile
